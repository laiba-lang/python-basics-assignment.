```python
menu = {
"tea": 120,
"coffee": 200,
"sandwich": 450,
"cookie": 150,
"cake_slice": 320
}
inventory = {
"tea": 40,
"coffee": 35,
"sandwich": 20,
"cookie": 50,
"cake_slice": 25
}
orders = [
{"tea": 2, "cookie": 1},
{"coffee": 1, "sandwich": 2},
{"tea": 3, "cake_slice": 1},
{"coffee": 2, "cookie": 2, "sandwich": 1},
{"tea": 1},
{"coffee": 1, "cake_slice": 2}
]
```

```python

for order in orders:
  for item in order:
    if item not in menu:
      pass
  processed_orders=[]
  for dish,qty in order.items():
    neworder={}
    if qty>inventory[dish]:
      qty=inventory[dish]
      inventory[dish] -= qty
    neworder[dish]=qty
  processed_orders.append(neworder)
  print(processed_orders)
```

```python
order_total=[]
for perorder in processed_orders:

  for dish, qty in perorder.items():
    total=qty*menu[dish]

    order_total.append(total)
print(order_total)
print(processed_orders)
```

```python
for order in processed_orders:
    for dish, qty in order.items():
        inventory[dish] -= qty  # subtract sold quantity

print("Updated Inventory:")
print(inventory)
```

```python

# day summary
revenue=sum(order_total)
print(revenue)
items_sold={}
# item sold
items_sold={}
for order in orders:
  for dish,qty in order.items():
    items_sold[dish]=items_sold.get(dish,0)+1
print(" the items sold are: ",items_sold)


```

```python
# top selling items:

max_qty = 0
most_sold_dish = ''

for dish, qty in items_sold.items():
    if qty > max_qty:
        max_qty = qty
        most_sold_dish = dish  # store the dish name

print(most_sold_dish)
```

```python
# 5. Low Stock Report:
for dish,qty in inventory.items():
  low_stock=""
  if qty<=5:
    print("low stock")
    low_stock=qty
```

```python
total_revenue=0
for order in orders:
    order_total = 0

    for dish, qty in order.items():
        price = menu[dish]
        items_total = price * qty
        order_total += items_total

    if order_total > 1000:
        final_total = order_total * 0.9
    else:
        final_total = order_total

    print(final_total)
    total_revenue += order_total

print("Total revenue (with discount) =", total_revenue, "PKR")
```

```python
 #7. Loyalty Points
order_points = []
total_points = 0

for order in orders:
    order_total = 0
    for dish, qty in order.items():
        price = menu[dish]
        item_total = price * qty
        order_total += item_total
    if order_total > 1000:
        discount = order_total * 0.10
        order_total_after_discount = order_total - discount
    else:
        order_total_after_discount = order_total
    points = int(order_total_after_discount // 100)
    order_points.append(points)
    total_points += points
print("Points for each order:", order_points)
print("Total points given today:", total_points)
```
